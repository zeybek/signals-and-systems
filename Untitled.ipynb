```matlab
clc; clear; close all;
syms t tau
x(t) = heaviside(t) - heaviside(t - 1); % input signal
h(t) = 7 * dirac(t) - 8 * exp(-1.4 * t) * heaviside(t); % impulse response
m = subs(x, tau) * subs(h, t - tau); % obtain x(tau)∗ h(t-tau)
y = int(m, tau, -inf, inf);
eval(y(0.7))

subplot(3, 1, 1);
ezplot(x, [-0.1, 2]);
title('Input Signal For LTI System');
xlabel('t')
ylabel('x(t)')
grid on;

subplot(3, 1, 2);
ezplot(h, [-0.1, 2]);
ylim([0 4.1])
title('LTI System Impulse Response');
xlabel('t')
ylabel('h(t)')
grid on;

subplot(3, 1, 3);
ezplot(y, [-0.1, 2]);
title('Result');
xlabel('t')
ylabel('y(t)')
grid on;
```
